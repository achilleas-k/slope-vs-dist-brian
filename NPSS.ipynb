```python
%pylab inline
import spikerlib as sl
import brian as bn
import gc
import random as rnd
```

```python
def get_Vws(trace, output_spikes, w):
    dt = bn.defaultclock.dt
    ws = output_spikes[0]-w
    ws_dt = (ws/dt).astype("int")
    return trace[0][ws_dt]

def calc_Vws(input_spikes, output_spikes, w, DV, tau):
    dt = bn.defaultclock.dt
    t0 = 0*ms
    Vws = []
    for outspk in output_spikes[0]:
        outspk = outspk*bn.second
        interval_inputs = [ti[1] for ti in input_spikes if (t0 <= ti[1]) and (ti[1] <= outspk-w)]
        cur_vws = sum([DV*exp(-(outspk-w-ii)/tau) for ii in interval_inputs])
        Vws.append(cur_vws)
        t0 = outspk
    return Vws
```

```python
bn.reinit()
bn.reinit_default_clock()
bn.clear(True)
gc.collect()
sim = bn.Network()
ms = bn.msecond
mV = bn.mvolt
Hz = bn.hertz
tau = 10*ms
Vth = 5*mV
DV = 3*mV
Vreset = 0*mV
w = 2*ms
neuron = bn.NeuronGroup(1, "dV/dt = -V/tau : volt", threshold="V>Vth", reset="V=Vreset")
input_spikes = []
pt = 0*ms
N = 4
for idx in range(100):
    pt += rnd.expovariate(10*Hz)
    input_spikes.append((idx % N, pt))
duration = max(input_spikes, key=lambda pair: pair[1])[1]+20*ms
input_group = bn.SpikeGeneratorGroup(N, input_spikes)
input_conn = bn.Connection(input_group, neuron, state='V', weight=DV)
voltage = bn.StateMonitor(neuron, 'V', record=True)
output_spikes = bn.SpikeMonitor(neuron)
sim.add(neuron, input_group, input_conn, voltage, output_spikes)
sim.run(duration)
voltage.insert_spikes(output_spikes, Vth*2)
plot(voltage.times, voltage[0])
plot(voltage.times, zeros_like(voltage.times)+Vth, 'k--')
print(output_spikes[0])
```

```python
actual_Vws = get_Vws(voltage, output_spikes, w)
theor_Vws = calc_Vws(input_spikes, output_spikes, w, DV, tau)
print(actual_Vws)
print(theor_Vws)
```
