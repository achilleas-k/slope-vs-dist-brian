### Generate spike trains using the synchrony parameters I use for NPSS but with very high jitter
I've discovered that jitter ($\sigma_{in}$) up to 4 ms does not affect the relationship between the input synchrony ($S_{in}$) and the multivariate spike distance.
Let's try higher values for ($\sigma_{in}$) and see what happens.

```python
def group_to_spiketrains(gengroup):
    # extract spiketimes from SpikeGeneratorGroup and convert to dictionary (can then use .items() to get arrays)
    spiketimes = gengroup.spiketimes
    spiketrains = {}
    for sp in spiketimes:
        if sp[0] in spiketrains:
            spiketrains[sp[0]].append(sp[1])
        else:
            spiketrains[sp[0]] = [sp[1]]
    for k in spiketrains.keys():
        spiketrains[k] = array(spiketrains[k])
    return spiketrains
```

```python
duration = 1.0
nsamples = int(duration/0.001)  # 1 sample per ms
Nin = 10
fin = 30
sync = frange(0, 1, 0.1)
sigma = frange(0, 0.01, 0.001)
kr_t = []
kr_d = []
for s in sync:
    for j in sigma:
        spikegen = sl.tools.fast_synchronous_input_gen(Nin, fin, s, j, duration)
        spiketrains = array(group_to_spiketrains(spikegen).values())
        t, d = sl.metrics.kreuz.multivariate(spiketrains, 0, duration, nsamples)
        kr_t.append(t)
        kr_d.append(d)
        print("{}, {} done".format(s, j), end="\r")
```

```python
kreuz = []
for t, d in zip(kr_t, kr_d):
    kreuz.append(trapz(d, t))
```

```python
plotsync = []
plotsigm = []
for s in sync:
    for j in sigma:
        plotsync.append(s)
        plotsigm.append(j)
        
```

```python
scatter(plotsync, kreuz, c=plotsigm)
cbar = colorbar()
title("Input synchrony $S_{in}$ vs Multivariate spike distance")
xlabel("$S_{in}$")
ylabel("MV SPIKE-d")
cbar.set_label("$\sigma_{in}$")
```

```python
scatter(plotsigm, kreuz, c=plotsync)
cbar = colorbar()
title("Input jitter $\sigma_{in}$ vs Multivariate spike distance")
xlabel("$\sigma_{in}$")
ylabel("MV SPIKE-d")
cbar.set_label("$S_{in}$")
```
