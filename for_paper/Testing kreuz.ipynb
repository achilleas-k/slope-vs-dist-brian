```python
import spikerlib as sl
kreuz = sl.metrics.kreuz.multivariate
```

```python
%pylab inline
```

```python
def gen_spiketrains(N, f, Sin, sigma, duration):
    N_sync = int(N*Sin)
    N_rand = N-N_sync
    spikearray = empty(N, dtype=object)
    for trainidx in range(0, N_rand):
        spikearray[trainidx] = sl.tools.poisson_times(f, duration)
    if N_sync:
        synctrain = sl.tools.poisson_times(f, duration)
        for trainidx in range(N_rand, N):
            spikearray[trainidx] = copy(synctrain)
            if sigma:
                jitter = random.normal(0, sigma, len(synctrain))
                spikearray[trainidx] += jitter
    return spikearray
```

- - - 
# All random

```python
N = 500
inrate = 10
duration = 5.0
spiketrains = gen_spiketrains(N, inrate, 0, 0, duration)
figure()
for h, sp in enumerate(spiketrains):
    plot(sp, zeros_like(sp)+h, 'b.')
xlabel("t")
ylabel("id#")
print("Mean rate: {} Hz".format(mean([len(st) for st in spiketrains])/duration))
dt = 0.001
samples = duration/dt
krt, krd = kreuz(spiketrains, 0, duration, samples)
print("SPIKE-distance: {}".format(trapz(krd, krt)))
```
